# Bar plot

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.dtypes
```

```python
#1 variables
ax = sns.countplot(x = 'sex', data = tips)
```

```python
#2 variabel
sns.set_theme(style = 'whitegrid')
ax = sns.barplot(x = "sex", y = "total_bill", data = tips)
```

```python
tips.info()
```

```python
tips.describe()
```

```python
from numpy import sum
ax = sns.barplot(x = "sex", y = "total_bill", data = tips, estimator = sum, ci = None)
```

```python
ax = sns.barplot(x="day", y = "total_bill", hue = "sex", data = tips)
```

```python
ax = sns.barplot(x = "tip", y = "day", data = tips)
```

```python
# mengubah urutan
ax = sns.barplot(x = "day", y = "tip", data = tips, order = ["Sat", "Thur", "Fri", "Sun"], estimator = sum, ci = None)
```

```python
ax = sns.barplot(x = "day", y = "tip", data = tips, order = temp.sort_values('tip')["day"], estimator=sum, ci = None, palette = "mako")
```
